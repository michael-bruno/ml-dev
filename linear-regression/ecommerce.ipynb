### Exploratory Data Analysis

_Time on Website_ vs _Yearly Amount Spent_

```python

sns.jointplot(x="Time on Website",y="Yearly Amount Spent",kind="scatter",data=customers)
```

_Time on App_ vs _Yearly Amount Spent_

```python

sns.jointplot(x="Time on App",y="Yearly Amount Spent",kind="scatter",data=customers)
```

_Time on App_ vs _Length of Membership_ (hex)

```python

sns.jointplot(x="Time on App",y="Length of Membership",kind="hex",data=customers)
```
